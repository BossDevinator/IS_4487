## Prepare Data

```python
df['total_amount'] = (df['product_price'] * df['quantity']).round(2)
df.head()
```

```python
df = df.dropna()
df.isnull().sum()
```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per month
- Display the total revenue for all months

```python
#average price
df.groupby('order_year_month')['product_price'].mean()
```

```python
#total revenue
df.groupby('order_year_month')['total_amount'].sum()
```
